# Hipotez Testi

Yolcuların ortalama yaşının 45 olduğunu varsayalım

Hipotez Testi Adımları

1. Hipotezlerin kurulması ve yönlerinin belirlenmesi
2. Anlamlılık düzeyi ve tablo değerinin belirlenmesi (alpha)
3. Test istatistiğinin belirlenmesi ve hesaplanması
4. Hesaplanan test istatistiği ile alfa'ya karşılık gelen tablo değerinin karşılaştırılması
5. Yorum

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import pylab
```

```python
# Veri seti içerisinden ilk 500 veriyi seçiyoruz.
data_age = data["Age"]
data_part_of_age =data_age[0:500]
```

```python
# odf DataFrame'inin istatistiksel özetini görüntüleyelim.
stats.describe(data_age)
```

Yukarıda görüldüğü gibi ortalama değer 39.379 olarak belirlenmiştir ve 45 değildir.

```python
data_age
```

```python
# Histogram
pd.DataFrame(data_age).plot.hist();
```

```python
# Qqplot
stats.probplot(data_age, dist="norm", plot=pylab)
pylab.show()
```

H0: Örnek dağılımı ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir fark yoktur

H1: Örnek dağılımı ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir fark vardır

Grafikteki noktalar bazı alanlarda kırmızı çizgi etrafında yoğunlaşmış fakat bazı bölgelerde dağılımın normal olmadığını söyleyebiliriz.

```python
from scipy.stats import shapiro
shapiro(data_part_of_age)
```

Fisher tarafından bu hatanın maksimum kabul edilebilir düzeyi 0,05 olarak önerilmiş ve kabul görmüştür.

Bir test sonucunda bulunan P-value 0,05'in altında bir değer ise karşılaştırma sonucunda anlamlı farklılık bulunduğu anlamına gelir.

p-value, bir örneklem verisinin normal bir dağılıma sahip olup olmadığını belirlemek için kullanılan önemli bir ölçüttür.


p-value değeri 0.05'den küçük olduğunda H0 hipotezini reddedebiliriz.


```python
print("T Hesap İstatistiği: " + str(shapiro(data_part_of_age)[0]))
print("Hesaplanan P-value: " + str(shapiro(data_part_of_age)[1]))
```

Burada p-value değeri 0.05'den büyük çıktığı için H0 hipotezini reddedemiyoruz.


Örnek dağılımı ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir fark yoktur

```python
stats.ttest_1samp(data_part_of_age, popmean = 45)
```

```python
from statsmodels.stats.descriptivestats import sign_test
sign_test(data_part_of_age, 170)
```

# Sonuç Modellerin Karşılaştırılması

```python
from sklearn.metrics import accuracy_score

models = {
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gaussian Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "CatBoost": CatBoostClassifier(silent=True),  # silence CatBoost output
}

for model_name, model in models.items():
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
      
    accuracy = accuracy_score(y_test,y_pred)
    
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy}")
    print()
      
   
    
   
```
